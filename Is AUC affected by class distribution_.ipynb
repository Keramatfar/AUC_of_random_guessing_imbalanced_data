```python
import numpy as np
np.random.seed(123)
from sklearn import metrics
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
aucs = []
ratio_of_zeros = []
for i in tqdm(range(100,10000)):
    imb_data = np.append(np.zeros(i), np.ones(100))  
    ratio_of_zeros.append(i/100)
    preds = np.random.choice(imb_data, size = len(imb_data))
    fpr, tpr, thresholds = metrics.roc_curve(imb_data, preds, pos_label=0)
    auc = metrics.auc(fpr, tpr)
    aucs.append(auc)
    #plt.plot(fpr, tpr, '--');
```

```python
plt.scatter(ratio_of_zeros, aucs);
```

```python
plt.hist(aucs, bins = 10);
```
